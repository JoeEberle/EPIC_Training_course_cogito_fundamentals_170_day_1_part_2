```python
definition = '''

## Designer UI

Dashboards and Dashboard Users

The **Designer UI** is used by both administrators and users.

Administrators use the Designer UI to edit the Original View of the source dashboard, which affects what all users see when they open the dashboard.

Users and Administrators use the Designer UI to edit a **personalized view** of a dashboard, which is visible
by default only to the user who created it.

#### Personalized View

When users modify a dashboard, their changes are saved on a user‐specific view of the dashboard. Most
users will have the security to personalize their view of any dashboard. This gives them a feeling of
ownership and helps improve adoption of, and trust in, dashboards. 

When a user or administrator edits a personalized view, they can only add components that have been distributed to them.

Personalization to dashboard views includes the following:

1. Add, remove, and rearrange components
2. Set defaults for dashboard parameters
3. Create and add a SlicerDicer component
4. Personalize components
    - Add, edit, rearrange, and delete links from link components
    - Change the display name or display color of a component
    - Add a badge to a table or graph component
    - Change the thresholds on a table or graph component

''' 
# Write the solution defitions out to the solution_description.md file
file_name = "solution_description.md"
with open(file_name, 'a',  encoding='utf-8') as f:
    # Write the template to the readme.md file
     f.write(definition)
        
display(Image(filename="radardashboards.png"))   

display(Markdown(definition))   
```

```python
definition = '''
##  Dashboard Editor

Administrators can create and edit source dashboards using the **Dashboard Editor**. There are a handful of
options that can only be accessed through the **Dashboard Editor**:

1. **Epic released dashboards** can only be edited in the Dashboard Editor. 
The system creates an override record for any changes made to an **Epic released dashboard**, such as adding or removing
components.
2. **Display name** vs **record name**: if useful, specify a display name that is distinct from the record
name. The record name will only be visible from the Dashboard Editor.
3. **Layouts**: Tripod, reverse tripod, and additional layouts are only available from the Dashboard Editor.
4. **Regions**: Components can be grouped into regions, and regions can be assigned custom names.
The  **Group Components** checkbox in the Dashboard Editor makes group names visible in the UI.
5. **Inactivation**: If a dashboard needs to be inactivated, this can be done from the Dashboard Editor.
Ready for use: If unchecked, the dashboard cannot be viewed, but can be edited in the
Dashboard Editor.
Enabled for user selection: If unchecked, the dashboard cannot be found in the Catalog, but
can be opened indirectly, such as with a link or with a redirector.

## Making it visible:

NOTE : Nobody can see your dashboard from the Analytics Catalog unless both the **Ready for use**
and **Enabled for user** selection checkboxes are selected

''' 
file_name = "solution_description.md" # Write the solution defitions out to the solution_description.md file
with open(file_name, 'a',  encoding='utf-8') as f:
     f.write(definition)
        
display(Markdown(definition))   
```
